```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp

# Constants
c1 = 1e-18
c2 = 0.1
c3 = 1e9
T = 2.0

#2.2

f_min = c3 / T
f = np.linspace(f_min * 1.001, 3e9, 1000)

E = c1 * f**2 + c2 / (T - c3 / f)

plt.plot(f, E)
plt.xlabel("f")
plt.ylabel("E(f)")
plt.title("2. Total Energy vs CPU Frequency")
plt.grid(True)
plt.show()

#2.3

# Positive variables for GP
f = cp.Variable(pos=True, name="f")
t_tx = cp.Variable(pos=True, name="t_tx")

# Objective: minimize total energy
objective = cp.Minimize(c1 * f**2 + c2 / t_tx)

# Constraint: latency deadline
constraints = [
    c3 / f + t_tx <= T
]

# Solve geometric program
problem = cp.Problem(objective, constraints)
problem.solve(gp=True)

# Results
print(f"Optimal transmission time t_tx* = {t_tx.value:.3f} s")
print(f"Minimum total energy = {problem.value:.3f} J")
print(f"Optimal CPU frequency f* = {f.value / 1e9:.3f} GHz")
```

## Question 3 (Aloha communication protocol) (Mark: 3 + 15 + 2 = 20 points)

**Aloha** is a well-known random access or _MAC_ (Media/multiple Access Control) communication protocol. In this system, multiple independent transmitters send packets (called $\textit{frames}$) to a single receiver over a shared channel. In the original Aloha, each transmitter sends whenever it has data, without coordinating with others. If two or more packets overlap in time, a collision occurs and the packets are lost; the transmitters then retry after a random delay.

In a more advanced setup, called **slotted Aloha**, the nodes can only transmit at the beginning of time slots, which are kept by a global/shared clock. The operations of slotted ALOHA in each node are decribed as follows:

* If there is only one frame (namely only one active transmitter) in a slot, then the transmission is successful and the slot is said to be a successful slot.
* If two or more frames (namely two or more active transmitters) collide in a slot, then the transmission is failed and a re-transmission is considered for each frame involved in the collision. The transmitter will retransmit its frame in each subsequent slot with probability $p$ until the frame is transmitted without a collision, where $0 \leq p \leq 1$.

Assume that there are $N$ transmitters and each transmitter independently attempts to transmit a frame in each slot with probability $p$. For one slot, let $S$ denote the _success probability_ that this slot is a successful slot.

Answer the following questions:

3. When $N$ tends to infinity, what is the maximal success probability $S$?

1.
For each transmitter, the probability of sending a successful signal is p, and the probability of not sending a signal of 1-p. Therefore, if there is to be a successful signal sent for a given slot, requiring N-1 sots to not transmit, is $p(1-p)^{N-1}$, with N slots, we get binomial: $Np(1-p)^{N-1}$.

2.

\begin{aligned}
\max_{p} \quad & S(p)=Np(1-p)^{N-1} \\
\text{subject to} \quad & 0 \le p \le 1, N \ge 1
\end{aligned}

With objective variable $p$ and decision variable $S(p)$.

Because this is a 1-D problem, we can determine convexity by the second derivative of the decision variable.

WE can then differentiate using the product rule:

$$
S'(p)
= N(1-p)^{N-1} + Np(N-1)(1-p)^{N-2}(-1)
$$

Factor out $N(1-p)^{N-2}$:

$$
S'(p)=N(1-p)^{N-2}\big[(1-p)-p(N-1)\big]
$$

Simplify the bracket:

$$
(1-p)-p(N-1)=1-Np
$$

So

$$
S'(p)=N(1-p)^{N-2}(1-Np)
$$

Differentiate again:

$$
S''(p)=\frac{d}{dp}\left[N(1-p)^{N-2}(1-Np)\right]
$$

Using the product rule:

$$
S''(p)
= N\left[-(N-2)(1-p)^{N-3}(1-Np)-N(1-p)^{N-2}\right]
$$

Factor out $N(1-p)^{N-3}$:

$$
S''(p)=N(1-p)^{N-3}\left[-(N-2)(1-Np)-N(1-p)\right]
$$

Now simplify the bracket:

$$
-(N-2)(1-Np)-N(1-p)
$$

$$
= -(N-2)+N(N-2)p-N+Np
$$

$$
= -2N+2+N(N-1)p
$$

Hence

$$
S''(p)=N(1-p)^{N-3}\left[N(N-1)p-2(N-1)\right]
$$

Factor out $(N-1)$:

$$
S''(p)=N(N-1)(1-p)^{N-3}(Np-2)
$$

We have

$$
S''(p)=N(N-1)(1-p)^{N-3}(Np-2)
$$

Since $N(N-1)(1-p)^{N-3} \ge 0$ on the feasible set, the sign depends on $Np-2$:

- $S''(p)<0$ when $p<\dfrac{2}{N}$
- $S''(p)>0$ when $p>\dfrac{2}{N}$

Therefore, the objective is:

- concave for $p<\dfrac{2}{N}$
- convex for $p>\dfrac{2}{N}$

So it is neither globally convex nor globally concave on $[0,1]$.

Therefore, this is not a convex optimisation problem.

We must then find the roots of the first derivative to find the optimal p value.

$$
S'(p)=N(1-p)^{N-2}(1-Np)=0
$$

This gives

$$
1-Np=0
$$

so

$$
p^*=\frac{1}{N}
$$

Now check the boundary points:

$$
S(0)=0
$$

$$
S(1)=N(1)(0)^{N-1}=0 \qquad (N>1)
$$

At the stationary point:

$$
S\left(\frac{1}{N}\right)
=
N\left(\frac{1}{N}\right)\left(1-\frac{1}{N}\right)^{N-1}
=
\left(1-\frac{1}{N}\right)^{N-1}
$$

Also,

$$
S''\left(\frac{1}{N}\right)
=
N(N-1)\left(1-\frac{1}{N}\right)^{N-3}(1-2)
<0
$$
so $p^*=\frac{1}{N}$ is a local maximum. Since the boundary values are both zero, it is also the global maximum.

The optimisation problem is

$$
\begin{aligned}
\max_{0 \le p \le 1} \quad & Np(1-p)^{N-1}
\end{aligned}
$$

The decision variable is

$$
p
$$

The objective is neither globally convex nor globally concave on $[0,1]$.

The optimality condition is

$$
S'(p)=N(1-p)^{N-2}(1-Np)=0
$$

which gives the optimal transmission probability

$$
p^*=\frac{1}{N}
$$

The maximum success probability is

$$
S(p^*)=\left(1-\frac{1}{N}\right)^{N-1}
$$

3.
Using the equation from 2, we can find

$$
\lim_{N\to\infty} S_{\max}
=
\lim_{N\to\infty}\left(1-\frac{1}{N}\right)^{N-1}
$$

This gives:
$$
\lim_{N\to\infty}\left(1-\frac{1}{N}\right)^{N-1}
=
\lim_{N\to\infty}\frac{\left(1-\frac{1}{N}\right)^N}{1-\frac{1}{N}}
$$

$$
\lim_{N\to\infty}1-\frac{1}{N} = 1
$$

So, the maximum probability of success is:

$$
\lim_{N\to\infty}\left(1-\frac{1}{N}\right)^N=e^{-1}

## Question 4 (Geometric programming: power control in wireless network) (Mark: 3 + 7 + 6 + 4 = 20 points)

### Example: Power Control in Wireless Communication

 *Adapted from Boyd, Kim, Vandenberghe, and Hassibi,* "[A Tutorial on Geometric Programming](https://web.stanford.edu/~boyd/papers/pdf/gp_tutorial.pdf)."

The [power control problem in wireless communications](http://winlab.rutgers.edu/~narayan/PAPERS/PC%20for%20Wireless%20Data.pdf) aims to minimise the total transmitter power available across $N$ trasmitters while concurrently achieving good (or a pre-defined minimum) performance.

The technical setup is as follows. Each transmitter $i$ transmits with a power level $P_i$ bounded below and above by a minimum and maximum level. The power of the signal received from transmitter $j$ at receiver $i$ is $G_{ij} P_{j}$, where $G_{ij} > 0$ represents the path gain (often loss) from transmitter $j$ to receiver $i$. The signal power at the intended receiver $i$ is $G_{ii} P_i$, and the interference power at receiver $i$ from other transmitters is given by $\sum_{k \neq i} G_{ik}P_k$. The (background) noise power at receiver $i$ is $\sigma_i$. Thus, the _Signal to Interference and  Noise Ratio (SINR)_ of the $i$th receiver-transmitter pair is

$$ S_i = \frac{G_{ii}P_i}{\sum_{k \neq i} G_{ik}P_k + \sigma_i }. $$

The minimum SINR represents a performance lower bound for this system, $S^{\text min}$.

The resulting optimisation problem is formulated as

$$
\begin{array}{ll}
\min_{P} & \sum_{i=1}^N P_i \\
\text{subject to} & P^{min} \leq P_i \leq P^{max}, \; \forall i \\
& \dfrac{G_{ii}P_i}{\sigma_i + \sum_{k \neq i} G_{ik}P_k} \geq S^{min} , \; \forall i \\
\end{array}
$$

```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set problem variables
N = 10

P_min = 0.1
P_max = 5
sigma = np.ones(N)*0.2
# Random path matrix generation created with Claude
G = np.random.uniform(0.1, 0.9, size=(N, N))
np.fill_diagonal(G, 1.0)
```

```python
def geo_solve(N, P_min, P_max, sigma, G, S_min):
  P = cp.Variable(N, pos=True)
  objective = cp.Minimize(cp.sum(P))

  constraints = []
  for i in range(N):
    # Broke interference sum and sigma into sep
    interference = cp.sum([S_min/G[i, i]*(G[i, k] *P[k]/P[i]) for k in range(N) if k != i])
    noise = (S_min*sigma[i])/(G[i,i]*P[i])
    constraints.append(interference + noise <= 1)

    constraints.append(P[i]>=P_min)
    constraints.append(P[i]<=P_max)

  prob = cp.Problem(objective, constraints)

  prob.solve(gp=True)
  plt.rcParams['axes.grid'] = True
  plt.rcParams['grid.alpha'] = 0.4
  plt.rcParams['figure.dpi'] = 108

  if prob.status in ['optimal', 'optimal_inaccurate']:
    SINR = []
    for i in range(N):
      signal = G[i, i]*P.value[i]
      interference = cp.sum([S_min/G[i, i]*(G[i, k] *P.value[k]/P.value[i]) for k in range(N) if k != i])
      noise = (S_min*sigma[i])/(G[i,i]*P.value[i])
      SINR.append(signal / (interference + noise))

    SINR = np.array(SINR)

    # Plotting generated by Claude
    palette = sns.color_palette('tab10', N)
    colors  = [palette[i] for i in range(N)]   # one colour per bar, not ambiguous

    links = [f'Link {i}' for i in range(N)]

    fig, axes = plt.subplots(1, 2, figsize=(11, 4))
    fig.suptitle(f'Power Control Results  (S_min = {S_min},  N = {N})', fontsize=13)

    # Plotting generated by Claude
    ax = axes[0]
    bars = ax.bar(links, P.value, color=colors, edgecolor='white', linewidth=0.8)
    ax.axhline(P_min, color='steelblue', linestyle='--', linewidth=1.2, label=f'P_min = {P_min}')
    ax.axhline(P_max, color='tomato',    linestyle='--', linewidth=1.2, label=f'P_max = {P_max}')
    # for bar, val in zip(bars, P.value):
    #     ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
    #             f'{val:.3f}', ha='center', va='bottom', fontsize=9)
    ax.set_title('Optimal transmit powers')
    ax.set_ylabel('Power')
    ax.set_ylim(0, np.max(P.value) * 1.2)
    ax.legend(fontsize=8)
    ax.set_xticks(range(N))
    ax.set_xticklabels(links, rotation=45, ha='right')

    # Plotting generated by Claude
    ax = axes[1]
    bars = ax.bar(links, SINR, color=colors, edgecolor='white', linewidth=0.8)
    ax.axhline(S_min, color='crimson', linestyle='--', linewidth=1.5, label=f'S_min = {S_min}')
    for bar, val in zip(bars, SINR):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
                f'{val:.3f}', ha='center', va='bottom', fontsize=9)
    ax.set_title('Achieved SINR per link')
    ax.set_ylabel('SINR')
    ax.set_ylim(0, max(SINR) * 1.3)
    ax.legend(fontsize=8)
    ax.set_xticks(range(N))
    ax.set_xticklabels(links, rotation=45, ha='right')

  print("S_min for solve:", S_min)
  print(f"Solver status : {prob.status}")
  print(f"Total power   : {prob.value:.6f}")
  print(f"Optimal powers: {P.value}")
  print()
```

```python
# S_min must be greater than  0 for any sort of meaningful answer other than P_i = P_min and breaks GP formulation as it isn't a valid monomial coeff
# therefore chose a number approx = 0
S_min = 0.0001
geo_solve(N, P_min, P_max, sigma, G, S_min)

S_min = 0.2
geo_solve(N, P_min, P_max, sigma, G, S_min)

# Max optimal value for S_min before infeasability
S_min = 0.21
geo_solve(N, P_min, P_max, sigma, G, S_min)

S_min = 0.3
geo_solve(N, P_min, P_max, sigma, G, S_min)
```

Initially, using a sufficiently small value of $S_{min}$ instead of  $S_{min}=0$ as the intial geometric solve function uses gp (geometric problem) = True, which forces a check for the function to be geometric. When $S_{min} = 0$ the function becomes linear and no longer meets the requirements to be geometric (posynomial and monomial coefficients > 0). This $S_{min}$ value essentially gives no lower bound requirement and thus all P values can be equal to $P_{min}$.

As the $S_{min}$ value is increased, the minimisation problem is no longer trivial and thus the optimal P values are no longer the minimum and increase in order to meet the minimum SINR values.

The last $S_{min}$ value that consistently is solvable regardless of the randomized G values is $S_{min} = 2.1$, however depending on the G values it was found to be able to get slightly larger ($S_{min} < 0.22$). Once the $S{min}$ value got larger though, the problem became infeasable to solve. This was surprisingly smaller than expected given that the P values were significantly smaller than $P_{max}$. This, however, is caused by how the randomised G matrix values were generated, as the  off-diagonal could be as large as 0.9. With a G value of 1 and 9 other terms that could be 0.9, the potential interference could be as large as 40.5, which when considered with $P_{max} = 5$ would have a $SINR ≈ 0.12$ which is well below $S_{min} = 0.22$. This explains such a low $S_{min}$ for feasibility and also means that as N increases with a randomised G matrix with off-diagnals up to 0.9 the possible $S_{min}$ value will also decrease unless the $P_{max}$ is powerful enough to compensate.

```python
def new_geo_solve(N, P_min, P_max, sigma, G, S_min):
  P = cp.Variable(N, pos=True)
  objective = cp.Minimize(cp.sum(P))

  constraints = []
  for i in range(N):
    interference = cp.sum([S_min/G[i, i]*0.5*G[i, j]*P[j]*cp.sum([(G[i, k] *P[k]/P[i]) for k in range(N) if (k != i and k != j)]) for j in range(N) if j != i])
    noise = (S_min*sigma[i])/(G[i,i]*P[i])
    constraints.append(interference + noise <= 1)

    constraints.append(P[i]>=P_min)
    constraints.append(P[i]<=P_max)

  prob = cp.Problem(objective, constraints)

  prob.solve(gp=True)
  print("S_min for solve:", S_min)
  print(f"Solver status : {prob.status}")
  print(f"Total power   : {prob.value:.6f}")
  print(f"Optimal powers: {P.value}")
  print()
```

### Background
Suppose we are given a binary classification dataset consisting of $n$ independent samples $S=\{(\mathbf{x}_i,y_i)\}^n_{i=1}$, where $\mathbf x_i∈\mathbb R^d$ represents the feature vector of the $i$-th sample and $y_i∈\{0,1\}$ is the corresponding label.


Logistic regression models the probability of the event $Y=1$ given the feature vector $\mathbf x$. Specifically, logistic regression assumes a linear relationship between the features $\mathbf{x}$ and the log-odds of the event $Y = 1$:

$$
\log \frac{p(y = 1|\mathbf{x})}{1 - p(y = 1|\mathbf{x})} = \mathbf{x}^\top \mathbf{w},
$$
where $\mathbf w\in \mathbb R^d$ is the weight vector.

Given the dataset, we estimate the weight vector $\mathbf w$ by minimizing the empirical risk under the log-loss (also known as cross-entropy loss):

$$
\hat{R}(\mathbf{w}) = - \frac{1}{n} \sum_{i = 1}^{n} \ell_\mathrm{log}(y_i, \mu_i(\mathbf{w})),
$$

where $\ell_\mathrm{log}(y, \mu) = y \log \mu + (1 - y) \log (1 - \mu)$.
